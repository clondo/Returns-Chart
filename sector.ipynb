## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Define a function to calculate RSI
def rsi(data, period):
    # Calculate the difference between consecutive prices
    delta = data.diff()
    # Drop the first row with NaN
    delta = delta.dropna()
    # Separate positive and negative changes
    up = delta.copy()
    down = delta.copy()
    up[up < 0] = 0
    down[down > 0] = 0
    # Calculate the exponential moving averages of positive and negative changes
    avg_gain = up.ewm(com=period-1, adjust=False).mean()
    avg_loss = down.abs().ewm(com=period-1, adjust=False).mean()
    # Calculate the relative strength
    rs = avg_gain / avg_loss
    # Calculate the RSI
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Define the market sectors and their corresponding SPDR ETFs
sectors = ["Communication Services", "Consumer Discretionary", "Consumer Staples", "Energy", "Financials", "Health Care", "Industrials", "Information Technology", "Materials", "Real Estate", "Utilities"]
etfs = ["XLC", "XLY", "XLP", "XLE", "XLF", "XLV", "XLI", "XLK", "XLB", "XLRE", "XLU"]

# Define the start and end dates for the data
start = "2023-04-01"
end = "2023-05-08"

# Define the period for RSI calculation (14 days by default)
period = 14

# Create an empty dataframe to store the RSI values for each sector
rsi_df = pd.DataFrame()

# Loop through each sector and ETF
for sector, etf in zip(sectors, etfs):
    # Get the price data for the ETF from yfinance
    data = yf.download(etf, start=start, end=end)
    # Calculate the RSI for the ETF using the function defined above
    rsi_data = rsi(data["Close"], period)
    # Add the RSI values to the dataframe with the sector name as column name
    rsi_df[sector] = rsi_data

# Plot the RSI values for each sector on a line chart
plt.figure(figsize=(12,8))
plt.plot(rsi_df)
plt.legend(sectors)
plt.title("Relative Strength Index of US Stock Market Sectors")
plt.xlabel("Date")
plt.ylabel("RSI")
plt.ylim(0,100)
# Add horizontal lines at 30 and 70 to indicate oversold and overbought levels
plt.axhline(30, color="red", linestyle="--")
plt.axhline(70, color="green", linestyle="--")
# Add a colorbar to show the relative strength of each sector based on RSI value
# Green means strong and red means weak
norm = plt.Normalize(rsi_df.iloc[-1].min(), rsi_df.iloc[-1].max())
sm = plt.cm.ScalarMappable(cmap="RdYlGn", norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label("Relative Strength")
plt.show()
```
